

$T=\frac{1}{2}m\left[\dot{r}^{2}+\left(r+a\right)^{2}\dot{\theta}^{2}\right]$

$V=\frac{1}{2}kr^{2}-mg\left(r+a\right)\cos\theta$

$L=\frac{1}{2}m\left[\dot{r}^{2}+\left(r+a\right)^{2}\right]\dot{\theta}^{2}-\frac{1}{2}kr^{2}+mg\left(r+a\right)\cos\theta$

Therefore, we can write down the Eular Lagrange equation

$m\ddot{r}+kr-mg\cos\theta-m\left(r+a\right)\dot{\theta}^{2}=0$

$m\left[2\left(r+a\right)\dot{r}\dot{\theta}+\left(r+a\right)^{2}\ddot{\theta}\right]+mg\left(r+a\right)\sin\theta=0$

At the equilibrium, $\theta=0$or $\theta=\pi$. When $\theta=0$,

$kr-mg=0$

$m\ddot{r}+kr=mg$

$r=A\cos\left(\sqrt{\frac{k}{m}}t\right)+B\sin\left(\sqrt{\frac{k}{m}}t\right)+\frac{mg}{k}$

Its corresponding frequency is

$\sqrt{\frac{k}{m}}$

This solution is stable.

When $\theta=\pi$

$kr+mg=0$

$r=A\cos\left(\sqrt{\frac{k}{m}}t\right)+B\sin\left(\sqrt{\frac{k}{m}}t\right)-\frac{mg}{k}$

Its corresponding frequency is

$\sqrt{\frac{k}{m}}$

This solution is very unstable. If $\theta$ is changed a little bit, the system will be out of the equilibrium.   

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(y, t, km, a):
    g=9.8
    return y[1], -km*y[0]+g*np.cos(y[2])+(y[0]+a)*y[3]**2, y[3], -(2*y[1]*y[3]+g*np.sin(y[2]))/(y[0]+a)
```

```python
y_0 = np.array([0.0, 0.0, 0.0, 0.0])
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(1,15))
plt.plot(t, y_sol[:,0],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
plt.show()
```

```python
y_0 = np.array([0.0, 0.0, 0.0+0.00001, 0.0])
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(1,15))
plt.plot(t, y_sol[:,0],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
plt.show()
```

```python
y_0 = np.array([-15, 0.0, np.pi, 0.0])
t = np.linspace(0, 20.0, 100000)
y_sol = odeint(f_int, y_0, t, args=(1,20))
plt.plot(t, y_sol[:,0],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
plt.show()
```

```python
y_0 = np.array([-15, 0.0, np.pi-0.00000001, 0.0])
t = np.linspace(0, 40.0, 1000000)
y_sol = odeint(f_int, y_0, t, args=(1,20))
plt.plot(t, y_sol[:,0],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
plt.show()
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(y_sol[:,2], y_sol[:,0])
#ax.set_rmax(5)
#ax.set_rticks([1.0, 2.0, 3.0, 4.0])  # less radial ticks
ax.grid(True)
plt.show()
```
