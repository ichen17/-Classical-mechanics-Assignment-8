Assumed that the displacement of five objects are $x_{1}$, $x_{2}$, $x_{3}$, $x_{4}$and $x_{5}$, we can write down the Largrange equation

$T=\frac{1}{2}m\left(\dot{x}_{1}^{2}+\dot{x}_{3}^{2}+\dot{x}_{5}^{2}\right)+\frac{1}{2}M\left(\dot{x}_{4}^{2}+\dot{x}_{5}^{2}\right)$

$V=\frac{1}{2}k\left(x_{2}-x_{1}\right)^{2}+\frac{1}{2}k\left(x_{3}-x_{2}\right)^{2}+\frac{1}{2}k\left(x_{4}-x_{3}\right)^{2}+\frac{1}{2}k\left(x_{5}-x_{4}\right)^{2}$

$L=\frac{1}{2}m\left(\dot{x}_{1}^{2}+\dot{x}_{3}^{2}+\dot{x}_{5}^{2}\right)+\frac{1}{2}M\left(\dot{x}_{4}^{2}+\dot{x}_{5}^{2}\right)-\frac{1}{2}\left[k\left(x_{2}-x_{1}\right)^{2}+k\left(x_{3}-x_{2}\right)^{2}+k\left(x_{4}-x_{3}\right)^{2}+k\left(x_{5}-x_{4}\right)^{2}\right]$

Therefore, we can write down the Eular Lagrange equation

$M\ddot{x}_{1}-k\left(x_{2}-x_{1}\right)=0$

$m\ddot{x}_{2}+k\left(x_{2}-x_{1}\right)-k\left(x_{3}-x_{2}\right)=0$

$M\ddot{x}_{3}+k\left(x_{3}-x_{2}\right)-k\left(x_{4}-x_{3}\right)=0$

$m\ddot{x}_{4}+k\left(x_{4}-x_{3}\right)-k\left(x_{5}-x_{4}\right)=0$

$M\ddot{x}_{5}+k\left(x_{5}-x_{4}\right)=0$

Assumed that the solution has form of $e^{i\omega t}$, the equation can be written down as

${\omega^{2}x_{1}+\frac{k}{M}\left(x_{2}-x_{1}\right)=0}$

$\omega^{2}x_{2}-\frac{k}{m}\left(2x_{2}-x_{1}-x_{3}\right)=0$

$\omega^{2}x_{4}-\frac{k}{m}\left(2x_{4}-x_{3}-x_{5}\right)=0$

$\omega^{2}x_{5}-\frac{k}{M}\left(x_{5}-x_{4}\right)=0$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(y, t, kM, km):
    
    return [y[1],kM*(y[2]-y[0]),y[3],km*(-2*y[2]+y[0]+y[4]),y[5],kM*(-2*y[4]+y[2]+y[6]),y[7],km*(-2*y[6]+y[4]+y[8]),y[9],kM*(y[6]-y[8])]
```

```python
y_0 = np.array([1.0, 0.0, 1.0, 0.0, 1.0, 0.0, 1.0, 0.0, 1.0, 0.0])
```

```python
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(1,np.sqrt(2)))
```

```python
plt.plot(t, y_sol[:,0],linewidth=1)
#plt.plot(t, y_sol[:,1],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
#plt.plot(t, y_sol[:,3],linewidth=1)
plt.plot(t, y_sol[:,4],linewidth=1)
plt.plot(t, y_sol[:,6],linewidth=1)
plt.plot(t, y_sol[:,8],linewidth=1)
plt.show()
```
