Assumed that the outer springs has the extention $x$, we can write down the length of springs

$l_{1}=a+x$

$l_{2}=l-2a-2x$

$l_{3}=a+x$

Because total length is $l$, we can find out $x$

$k_{0}\left(l-2a-2x-a_{0}\right)=kx$

$\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}=x$

$\frac{lk}{2k_{0}+k}-\frac{2ak}{2k_{0}+k}-\frac{a_{0}k}{2k_{0}+k}$

Assumed that the displacement of two springs are $x_{1}$and $x_{2}$, we can write down the Largrange equation

$T=\frac{1}{2}m\left(\dot{x}_{1}^{2}+\dot{x}_{2}^{2}\right)$

$V=\frac{1}{2}k\left(\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{1}\right)^{2}+\frac{1}{2}k_{0}\left[\frac{k\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{2}-x_{1}\right]^{2}+\frac{1}{2}k\left[\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}-x_{2}\right]^{2}$

$L=\frac{1}{2}m\left(x_{1}^{2}+x_{2}^{2}\right)-\frac{1}{2}k\left(\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{1}\right)^{2}-\frac{1}{2}k_{0}\left[\frac{k\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{2}-x_{1}\right]^{2}-\frac{1}{2}k\left[\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}-x_{2}\right]^{2}$

Therefore, we can write down the Eular Lagrange equation

$m\ddot{x}_{1}+k\left[\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{1}\right]-k_{0}\left[\frac{k\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{2}-x_{1}\right]=0$

$m\ddot{x}_{2}-k\left[\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}-x_{2}\right]+k_{0}\left[\frac{k\left(l-2a-a_{0}\right)}{2k_{0}+k}+x_{2}-x_{1}\right]=0$

Assumed that the solution of $x_{1}$ and $x_{2}$is $e^{i\omega t}$, the equations become

$\omega^{2}x_{1}+\left(\frac{k+k_{0}}{m}\right)x_{1}-\frac{k_{0}}{m}x_{2}=0$

$\omega^{2}x_{2}+\left(\frac{k+k_{0}}{m}\right)x_{2}-\frac{k_{0}}{m}x_{1}=0$

$\frac{1}{m}\left(\begin{array}{cc}
k+k_{0} & -k_{0}\\
-k_{0} & k+k_{0}
\end{array}\right)\left(\begin{array}{c}
x_{1}\\
x_{2}
\end{array}\right)=\omega^{2}\left(\begin{array}{c}
x_{1}\\
x_{2}
\end{array}\right)$

$\frac{k+k_{0}}{m}-\omega^{2}=\pm\frac{k_{0}}{m}$

$\omega=\sqrt{\frac{k}{m}}$

The correspond eigenstate is

$x'_{1}=\frac{1}{\sqrt{2}}\left(\begin{array}{c}
x_{1}\\
x_{2}
\end{array}\right)$

$\omega=\sqrt{\frac{k+2k_{0}}{m}}$

The correspond eigenstate is

$x'_{2}=\frac{1}{\sqrt{2}}\left(\begin{array}{c}
x_{1}\\
-x_{2}
\end{array}\right)$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

Let the variable 

$\frac{k_{0}\left(l-2a-a_{0}\right)}{2k_{0}+k}=\tilde{l}_{0}$

$\frac{k\left(l-2a-a_{0}\right)}{2k_{0}+k}=\tilde{l}$

$\frac{k}{m}=d$

$\frac{k_{0}}{m}=\frac{d\tilde{l}_{0}}{\tilde{l}}$

```python
#l0=2
#l=1
#d0=4
#d=2
def f_int(y, t, l0, l, d):
    d0=d*l0/l
    return [y[1],-d*(l0+y[0])+d0*(l+y[2]-y[0]),y[3],d*(l0-y[2])-d0*(l+y[2]-y[0])]
```

```python
y_0 = np.array([1.0, 0.0, 1.0, 0.0])
```

```python
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(2,1,2))
```

The mode the two objects vibrating in same direction: 

```python
plt.plot(t, y_sol[:,0],linewidth=1)
#plt.plot(t, y_sol[:,1],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
#plt.plot(t, y_sol[:,3],linewidth=1)
plt.show()
```

The mode the two objects vibrating in same direction: 

```python
y_0 = np.array([1.0, 0.0, -1.0, 0.0])
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(2,1,2))
plt.plot(t, y_sol[:,0],linewidth=1)
#plt.plot(t, y_sol[:,1],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
#plt.plot(t, y_sol[:,3],linewidth=1)
plt.show()
```

The combination of two modes:

```python
y_0 = np.array([0.5, 1.0, 0.5, -1.0])
t = np.linspace(0, 10.0, 10000)
y_sol = odeint(f_int, y_0, t, args=(2,1,2))
plt.plot(t, y_sol[:,0],linewidth=1)
#plt.plot(t, y_sol[:,1],linewidth=1)
plt.plot(t, y_sol[:,2],linewidth=1)
#plt.plot(t, y_sol[:,3],linewidth=1)
plt.show()
```
